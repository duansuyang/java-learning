```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Load the data
file_path = 'Currency_Exchange(updated).csv'
data = pd.read_csv(file_path)



data.info()

```

```python
print(data.describe())
```

```python
data["DateNew"] = pd.to_datetime(data["Date"], format="%d/%m/%Y")
data['DaysSinceStart'] = (data['DateNew'] - data['DateNew'].min()).dt.days

data['Year'] = data['DateNew'].dt.year
data["Month"] = data["DateNew"].dt.month

data["CAN_UK"] = data["Canadian Dollar"]/data["UK Pound"]
print(data.head())
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Create a figure with Plotly
figure = make_subplots(specs=[[{"secondary_y": False}]])

# Add Canadian Dollar exchange rate line
figure.add_trace(go.Scatter(x=data['DateNew'], y=data['Canadian Dollar'],
                            mode='lines', name='Canadian Dollar',
                            line=dict(color='blue')), secondary_y=False)

# Add UK Pound exchange rate line
figure.add_trace(go.Scatter(x=data['DateNew'], y=data['UK Pound'],
                            mode='lines', name='UK Pound',
                            line=dict(color='red')), secondary_y=False)

# Update figure layout
figure.update_layout(title='Exchange Rates: Canadian Dollar vs. UK Pound',
                     xaxis_title='Date',
                     yaxis_title='Exchange Rate')

figure.show()
```

```python
# 创建直方图
plt.hist(data["CAN_UK"], bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.show()
```

```python
# Random sampling
from sklearn.model_selection import train_test_split
# train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)
# 分割训练集和测试集
train_size = int(0.8 * len(data))  # 使用80%的数据作为训练集
train_set = data[:train_size]
test_set = data[train_size:]
```

```python
train_set
```

```python
from pandas.plotting import scatter_matrix
attributes = ["Canadian Dollar", "UK Pound", "DateNew","DaysSinceStart",	"Year", "Month","CAN_UK"]
scatter_matrix(data[attributes], figsize=(12,8))
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.impute import KNNImputer

num_pipeline = make_pipeline(
    KNNImputer(n_neighbors=3),
    StandardScaler())

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
)
```

```python
train_set.dropna(inplace=True)
CAN_UK = train_set[["DateNew","DaysSinceStart",	"Year", "Month"]]
CAN_UK_lable = train_set["CAN_UK"].copy()


test_set.dropna(inplace=True)
CAN_UK_test = test_set[["DateNew","DaysSinceStart",	"Year", "Month"]]
CAN_UK_test_lable = test_set["CAN_UK"].copy()
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = make_pipeline(preprocessing, LinearRegression())
lin_reg.fit(CAN_UK, CAN_UK_lable)
```

```python
CAN_UK_predictions = lin_reg.predict(CAN_UK_test)
CAN_UK_predictions[:5].round(-2)  # -2 = rounded to the nearest hundred
```

```python
from sklearn.metrics import mean_squared_error

lin_rmse = mean_squared_error(CAN_UK_test_lable, CAN_UK_predictions,
                              squared=False)
lin_rmse
```

```python
from sklearn.metrics import r2_score

lin_r2_score = r2_score(CAN_UK_test_lable, CAN_UK_predictions)
lin_r2_score
```

```python
CAN_UK_result = test_set.copy()
CAN_UK_result["Predictions"] = CAN_UK_predictions
# CAN_UK_result = CAN_UK_result.sort_values('DaysSinceStart')

CAN_UK_result = pd.concat([train_set, CAN_UK_result], ignore_index=True)

CAN_UK_result.head(10)
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Create a figure with Plotly
figure = make_subplots(specs=[[{"secondary_y": False}]])

# Add Canadian Dollar exchange rate line
figure.add_trace(go.Scatter(x=CAN_UK_result["DateNew"], y=CAN_UK_result["CAN_UK"],
                            mode='lines', name='Actual',
                            line=dict(color='blue')), secondary_y=False)

# Add UK Pound exchange rate line
figure.add_trace(go.Scatter(x=CAN_UK_result["DateNew"], y=CAN_UK_result["Predictions"],
                            mode='lines', name='Predicted',
                            line=dict(color='red')), secondary_y=False)

# Update figure layout
figure.update_layout(title='Exchange Rates of Canadian Dollar with UK Pound: Actual vs. Predicted',
                     xaxis_title='Date',
                     yaxis_title='Exchange Rate')

figure.show()
```
